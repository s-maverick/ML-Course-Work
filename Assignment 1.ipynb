```python
train = pd.read_csv('/kaggle/input/mlss-sp-25/train.csv')
test = pd.read_csv('/kaggle/input/mlss-sp-25/test.csv')
sample_submission = pd.read_csv('/kaggle/input/mlss-sp-25/sample_submission.csv')
object_id = test['object_id']
```

```python
col =['height_cm','width_cm', 'location','transaction_currency','category','creation_year','is_titled', 'signed', 'dated','n_exhibited', 'artist_birth_year', 'estimate']
train = train[col + ['price_realized']] 
test = test[col]
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
drop_col = ['creation_year','is_titled']
train = train.drop(drop_col, axis = 1)
test = test.drop(drop_col, axis = 1)
```

```python
def clean_numeric(value):
    try:
        return float(value)
    except:
        return None

train = train.copy()
test = test.copy()

train.loc[:, 'height_cm'] = train['height_cm'].apply(clean_numeric)
train.loc[:, 'width_cm'] = train['width_cm'].apply(clean_numeric)
test.loc[:, 'height_cm'] = test['height_cm'].apply(clean_numeric)
test.loc[:, 'width_cm'] = test['width_cm'].apply(clean_numeric)
```

```python
mode_imputatipn_col = ['location', 'artist_birth_year']
from sklearn.impute import SimpleImputer
mode_imputer = SimpleImputer(strategy = 'most_frequent')
train[mode_imputatipn_col] = mode_imputer.fit_transform(train[mode_imputatipn_col])
test[mode_imputatipn_col] = mode_imputer.transform(test[mode_imputatipn_col])
```

```python
median_imputatipn_col = ['estimate', 'height_cm', 'width_cm']
from sklearn.impute import SimpleImputer
median_imputer = SimpleImputer(strategy = 'median')
train[median_imputatipn_col] = median_imputer.fit_transform(train[median_imputatipn_col])
test[median_imputatipn_col] = median_imputer.transform(test[median_imputatipn_col])
```

```python
train['area'] = train['height_cm'] * train['width_cm']
test['area'] = test['height_cm'] * test['width_cm']
drop_col = ['height_cm','width_cm']
train = train.drop(drop_col, axis = 1)
test = test.drop(drop_col, axis = 1)
```

```python
from datetime import datetime
current_year = datetime.now().year
train['aprox_age_painting'] = ((current_year - train['artist_birth_year'])/10).astype(int)*10
test['aprox_age_painting'] = ((current_year - test['artist_birth_year'])/10).astype(int)*10
train = train.drop('artist_birth_year', axis = 1)
test = test.drop('artist_birth_year', axis = 1)
```

```python
catagorical_cols = ['category', 'signed', 'dated', 'location', 'transaction_currency']
#catagorical_cols = ['category', 'signed', 'dated', 'n_exhibited', 'aprox_age_painting']
train_dummies = pd.get_dummies(train[catagorical_cols], drop_first = True).astype(int)
test_dummies = pd.get_dummies(test[catagorical_cols], drop_first = True).astype(int)
```

```python
test_dummies = test_dummies.reindex(columns = train_dummies.columns, fill_value = 0)
```

```python
test_dummies
```

```python
train
```

```python
train = pd.concat([train.drop(catagorical_cols, axis = 1), train_dummies], axis = 1)
test = pd.concat([test.drop(catagorical_cols, axis = 1), test_dummies], axis = 1)
```

```python
train
```

```python
from sklearn.linear_model import LinearRegression
X_train = train.drop('price_realized', axis = 1)
y_train = train['price_realized']
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
pred = lr_model.predict(test)
```

```python
sample_submission
```

```python
submission = pd.DataFrame()
submission['object_id'] = object_id
submission['price_realized1'] = pred
submission
```

```python
submission.to_csv('submission.csv', index = False)
```

```python
import pandas as pd
import numpy as np

def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

actuals = sample_submission
submission = submission

df = submission.merge(actuals, on="object_id")

mape = mean_absolute_percentage_error(df["price_realized"], df["price_realized1"])
print(f"MAPE: {mape:.2f}%")
```

```python
import statsmodels.api as sm
# Add constant term for intercept
X_train = sm.add_constant(X_train)
test = sm.add_constant(test)

# Fit the linear regression model
model = sm.OLS(y_train, X_train).fit()

# Predict on the test set
predictions = model.predict(test)

# Display summary of model
print(model.summary())
```
